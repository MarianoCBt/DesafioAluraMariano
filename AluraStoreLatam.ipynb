### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda2.head()
```

# **📄 Informe Final – Análisis de Tiendas**
###**Introducción**

El objetivo de este análisis fue evaluar el desempeño de cuatro tiendas con el fin de recomendar al Sr. Juan cuál de ellas debería vender. Para ello, se analizaron distintos indicadores clave, tales como los ingresos totales, las categorías de productos más y menos vendidas, la calificación promedio de los clientes, los productos más y menos vendidos, y el costo de envío promedio por tienda.

A partir de estos datos y de las visualizaciones generadas, se buscó identificar fortalezas y debilidades de cada tienda para tomar una decisión fundamentada.

## **Desarrollo del análisis**
###**Ingresos totales**

El gráfico de ingresos totales por tienda muestra que Tienda 1 presenta los mayores ingresos, seguida por Tienda 2 y Tienda 3. Tienda 4 es la que registra el menor nivel de ingresos.
Este indicador sugiere que Tienda 1 tiene un mayor volumen de ventas y una mejor capacidad de generación de ingresos en comparación con las demás.

###**Calificación promedio de los clientes**

El gráfico de calificación media por tienda muestra que todas las tiendas presentan valoraciones similares, cercanas a 4 sobre 5. Sin embargo, Tienda 3 y Tienda 2 muestran una leve ventaja en términos de satisfacción del cliente, mientras que Tienda 1 y Tienda 4 se mantienen apenas por debajo.
Esto indica que, en general, el nivel de satisfacción es alto en todas las tiendas, sin diferencias extremadamente marcadas.

###**Costo de envío promedio**

En cuanto al costo de envío promedio, el gráfico evidencia que Tienda 1 presenta el costo de envío más alto, mientras que Tienda 4 tiene el costo más bajo. Tienda 2 y Tienda 3 se ubican en un punto intermedio.
Este factor es relevante, ya que un mayor costo de envío puede afectar la decisión de compra de los clientes, aunque en este caso el envío es pagado por el cliente.

###**Conclusión**

A partir de los análisis realizados y de las visualizaciones generadas, se puede concluir que la Tienda 4 es la menos eficiente dentro de la cadena Alura Store. Esta tienda presenta los ingresos totales más bajos en comparación con las demás, lo que evidencia un menor volumen de ventas y un rendimiento comercial inferior.

Si bien la calificación promedio de los clientes de la Tienda 4 es similar a la del resto de las tiendas, este factor no logra compensar su bajo desempeño en términos de ingresos. Además, aunque cuenta con el costo de envío promedio más bajo, este aspecto no se traduce en un mejor resultado económico ni en una ventaja competitiva clara.
